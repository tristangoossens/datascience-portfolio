```python
import seaborn as sns
import pandas as pd
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

## 3.1 Categorical data
The volgende categorische attributen is terug te vinden in de dataset 
- Species
- Island
- Sex

### 3.1.1 Sex ('Geslacht')

```python
penguins['sex'].unique()
```

```python
penguins['sex'].value_counts(dropna=False)
```

```python
penguins['sex'].value_counts(dropna=True).plot(kind='bar')
```

```python
penguins['sex'].value_counts(dropna=True).plot(kind='pie', autopct='%1.0f%%')
```

#### Observeringen

- Het geslacht van de penguins is gelijk verspreid over de dataset (M: 168, F: 165)
- Enkele records uit de dataset kennen geen geslacht (11)

### 3.1.2 Soort ("species")
Analyse op **enkel** de soort penguin

```python
penguins['species'].unique()
```

```python
penguins['species'].value_counts(dropna=False)
```

```python
penguins['species'].value_counts().plot(kind='bar')
```

```python
penguins['species'].value_counts(dropna=True).plot(kind='pie', autopct='%1.0f%%')
```

#### Observeringen

- De soort van de penguin kent een grotere spreiding dan het geslacht
- Elke record (penguin) in de dataset heeft een soort

### 3.1.3 Eiland ("island")
