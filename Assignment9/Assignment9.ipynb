```python
import pandas as pd
import seaborn as sns

penguins = sns.load_dataset("penguins")
```

## 9.1 Correlations penguin data set

```python
penguins.corr().style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

#### Observaties
- Een hoge positieve correlatie tussen body_mass_g (gewicht) en flipper_length_mm (lengte flipper) is logisch, bij een langere flipper zal een pinguin ook meer wegen
- Een sterke negatieve correlatie bestaat tussen flipper_length_mm (lengte flipper) en bill_depth_mm (diepte bek). Namelijk hoe kleiner de lengte van de flipper, hoe dieper de bek is

## 9.2 Strongest positive correlation
De "strongest positive correlation" ligt tussen de kolommen body_mass_g (Lichaamsgewicht) en flipper_length_mm (Lengte flipper in milimeter)

```python
sns.regplot(data=penguins, x='body_mass_g', y='flipper_length_mm', fit_reg=True, line_kws={"color": "red"})
```

## 9.2 Strongest negative correlation
De "Strongest negative correlation" ligt tussen de kolommen flipper_length_mm en bill_depth_mm

```python
sns.regplot(data=penguins, x='bill_depth_mm', y='flipper_length_mm', fit_reg=True, line_kws={"color": "red"})
```

## 9.3 Weakest correlation 
De zwakste correlatie ligt tussen de kolommen bill_depth_mm en bill_length_mm

```python
sns.regplot(data=penguins, x='bill_depth_mm', y='bill_length_mm', fit_reg=True, line_kws={"color": "red"})
```
