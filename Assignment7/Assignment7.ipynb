## 7.1 White rating distribution
In assignment 6 is er een univariate analyse uitgevoerd op de gecombineerde rating van beide spelen (rating zwart + rating wit). Nu word de distributie bepaald voor enkel de spelers met de witte stukken

```python
chess['white_rating'].plot(kind='hist', bins=15, title='Distribution rating white')
```

```python
chess['white_rating'].plot(kind='kde', title='Distribution turns')
```

### 7.1.1 Conclusie distributie
De distributie van de data van rating voor de witte speler is "Double Exponential Distribution".

## 7.2 Turns distribution
In assignment 6 heb ik een univariate analyse uitgevoerd op het aantal zetten die tijdens een pot worden gespeeld. Tijdens die analyse zijn de outliers van de dataset weggehaald. In deze assigment zal ik hetzelfde doen.

```python
withoutTurnOutliers = chess["turns"][chess["turns"] < 135]
withoutTurnOutliers.plot(kind='hist', bins=25, title='Aantal zetten zonder uitschieters (turns < 135) (boxplot)')
```

```python
withoutTurnOutliers.plot(kind='kde', title='Aantal zetten zonder uitschieters (turns < 135) (boxplot)')
```

### 7.2.1 Conclusie distributie
De distributie van de data van aantal zetten is: "Normal distribution". Wel zijn bij deze analyse de uitschieters uit de dataset gefilterd
