## 11.1 Numerical means grouped by species

```python
penguins.groupby('species').mean()
```

### 11.1.1 Observeringen
- Een Gentoo penguin is gemiddeld een stuk zwaarder dan de andere twee soorten penguins
- Een Adelie penguin heeft gemiddeld een kortere snavel dan de andere soorten
- Een Gentoo heeft een lange maar dunne snavel in vergelijking met de andere soorten
- Een Gentoo penguin heeft de langste flipper lengte van alle soorten

## 11.2 Barplot mean and confidence
