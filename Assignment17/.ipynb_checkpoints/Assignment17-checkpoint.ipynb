```python
import pandas as pd
import seaborn as sns
import numpy as np

# Load penguin dataset and remove missing values
penguins = sns.load_dataset("penguins").dropna()
```

```python
penguins.corr() # to find out which features correlate with body_mass_g
```

## 17.1 Split data

```python
from sklearn.model_selection import train_test_split

# Split dataset into a training (70%) and testing set (30%)
# Based on random 'seed' 42 and equally divides species across both datasets
penguinsTrainSet, penguinsTestSet = train_test_split(penguins, test_size=0.3, random_state=42, stratify=penguins['species'])
print(penguinsTrainSet.shape, penguinsTestSet.shape)
```

## 17.2 Train model

```python
from sklearn.tree import DecisionTreeRegressor

features= ['flipper_length_mm']
dt_regression = DecisionTreeRegressor(max_depth = 8) # Increase max_depth to see effect in the plot
dt_regression.fit(penguinsTrainSet[features], penguinsTrainSet['body_mass_g'])
```

## 17.3 Evaluate DecisionTreeRegressor

```python
def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)
```

```python
predictionsOnTrainset = dt_regression.predict(penguinsTrainSet[features])
predictionsOnTestset = dt_regression.predict(penguinsTestSet[features])

rmseTrain = calculate_rmse(predictionsOnTrainset, penguinsTrainSet['body_mass_g'])
rmseTest = calculate_rmse(predictionsOnTestset, penguinsTestSet['body_mass_g'])

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```

## 17.4 Plot DecisionTreeRegressor

```python
from sklearn import tree
import graphviz

def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
plot_tree_regression(dt_regression, features)
```
