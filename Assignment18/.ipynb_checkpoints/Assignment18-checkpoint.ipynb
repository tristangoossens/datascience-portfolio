```python
import pandas as pd
import seaborn as sns
import numpy as np

chess = pd.read_csv("../assets/chess.csv", sep=',')
chess['combined_rating'] = chess['black_rating'] + chess['white_rating']
```

## 16.1 Split data

```python
from sklearn.model_selection import train_test_split

# Split dataset into a training (70%) and testing set (30%)
# Based on random 'seed' 42 and equally divides species across both datasets
chessTrainSet, chessTestSet = train_test_split(chess, test_size=0.3, random_state=42, stratify=chess['victory_status'])
print(chessTrainSet.shape, chessTestSet.shape)
```

## 16.2 Train model

```python
chess.corr().style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

```python
from sklearn.tree import DecisionTreeRegressor

features= ['combined_rating']
dt_regression = DecisionTreeRegressor(max_depth=8) # Increase max_depth to see effect in the plot
dt_regression.fit(chessTrainSet[features], chessTrainSet['opening_ply'])
```

## 16.3 Evaluate DecisionTreeRegressor

```python
def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)
```

```python
predictionsOnTrainset = dt_regression.predict(chessTrainSet[features])
predictionsOnTestset = dt_regression.predict(chessTestSet[features])

rmseTrain = calculate_rmse(predictionsOnTrainset, chessTrainSet['opening_ply'])
rmseTest = calculate_rmse(predictionsOnTestset, chessTestSet['opening_ply'])

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```

## 17.4 Plot DecisionTreeRegressor

```python
from sklearn import tree
import graphviz

def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
plot_tree_regression(dt_regression, features)
```
